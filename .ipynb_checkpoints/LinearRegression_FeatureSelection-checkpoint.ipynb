### Feature selection and linear regression with diabetes dataset

Feature Selection is one of the core concepts in machine learning which hugely impacts the performance of model. The data features that are used to train machine learning models have a huge influence on the performance of model.
Irrelevant or partially relevant features can negatively impact model performance. The benefits of performing feature selection includes:
1. Reduces Overfitting: Less redundant data means less opportunity to make decisions based on noise.
2. Improves Accuracy: Less misleading data means modeling accuracy improves.
3. Reduces Training Time: fewer data points reduce algorithm complexity and algorithms train faster.

#### Feature selection (https://www.datacamp.com/community/tutorials/feature-selection-python)
1. It enables the machine learning algorithm to train faster.
2. It reduces the complexity of a model and makes it easier to interpret.
3. It improves the accuracy of a model if the right subset is chosen.
4. It reduces Overfitting.

There are three types of general feature selection methods - Filter methods, Wrapper methods, and Embedded methods.

##### Filter methods
<img src="filter_method.png">
Filter method relies on the general uniqueness of the data to be evaluated and pick feature subset, not including any mining algorithm. Filter method uses the exact assessment criterion which includes distance, information, dependency, and consistency. The filter method uses the principal criteria of ranking technique and uses the rank ordering method for variable selection. 

Filter methods are generally used as a data preprocessing step. The selection of features is independent of any machine learning algorithm. Features give rank on the basis of statistical scores which tend to determine the features' correlation with the outcome variable. You can refer to the following table for defining correlation coefficients for different types of data (in this case continuous and categorical).

<img src="correlation_for_different_data_types.png">

###### SelectKBest method
Select features according to the k highest scores. The score function for selecting features includes:
1. For regression: f_regression, mutual_info_regression
2. For classification: chi2, f_classif, mutual_info_classif

###### SelectKBest based on f_regression

This is done in 2 steps:
1. The correlation between each regressor and the target is computed, that is, ((X[:, i] - mean(X[:, i])) * (y - mean_y)) / (std(X[:, i]) * std(y)).
2. It is converted to an F score then to a p-value.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
fs = SelectKBest(f_regression, 'all')
fs.fit(x_train, y_train)
```

```python
plt.plot(x_train.columns, fs.scores_)
```

BMI, S5, Blood, S4, S3, and S6 might be selected for later analysis

```python
fs = SelectKBest(f_regression, 6)
fs.fit(x_train, y_train)
x_train_fs = fs.transform(x_train)
x_test_fs = fs.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train_fs, y_train) # same as lm.fit(x_train[["BMI", "S5", "Blood", "S4", "S3", "S6"]], y_train)
print(lm.intercept_, lm.coef_, lm.score(x_test_fs, y_test))
```

###### SelectKBest based on mutual_info_regression
Mutual information (MI) between two random variables is a non-negative value, which measures the dependency between the variables. It is equal to zero if and only if two random variables are independent, and higher values mean higher dependency.
The function relies on nonparametric methods based on entropy estimation from k-nearest neighbors distances.

```python
from sklearn.feature_selection import mutual_info_regression
fs = SelectKBest(mutual_info_regression, 'all')
fs.fit(x_train, y_train)
```

```python
plt.plot(x_train.columns, fs.scores_)
```

```python
fs = SelectKBest(mutual_info_regression, 6)
fs.fit(x_train, y_train)
x_train_fs = fs.transform(x_train)
x_test_fs = fs.transform(x_test)
lm = LinearRegression()
lm.fit(x_train_fs, y_train) # sligthly different from lm.fit(x_train[["BMI", "S5", "Blood", "S4", "S3", "S6"]], y_train)
print(lm.intercept_, lm.coef_, lm.score(x_test_fs, y_test))
```

##### Wrapper methods
A wrapper method needs one machine learning algorithm and uses its performance as evaluation criteria. This method searches for a feature which is best-suited for the machine learning algorithm and aims to improve the mining performance. 
<img src= "wrapper_method.png">

```python
###### Tuning k using CV (https://machinelearningmastery.com/feature-selection-for-regression-data/)
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
rkf = RepeatedKFold(n_splits = 10, n_repeats = 3, random_state = 0)
scores = []
for i in range(1, x_train.shape[1]):
    fs = SelectKBest(mutual_info_regression, i)
    lm = LinearRegression()
    pipeline = Pipeline(steps=[('sel', fs), ('lr', lm)])
    score = cross_val_score(pipeline, x_train, y_train, cv = rkf)
    scores.append(np.mean(score))
```

```python
plt.plot(range(1,  x_train.shape[1]), scores)
```

k = 6 seems to be good, where score reaches a plateau.

###### RFE
Some typical examples of wrapper methods are forward feature selection, backward feature elimination, recursive feature elimination. (https://scikit-learn.org/stable/auto_examples/feature_selection/plot_select_from_model_diabetes.html#sphx-glr-auto-examples-feature-selection-plot-select-from-model-diabetes-py)

Recursive Feature elimination: Recursive feature elimination performs a greedy search to find the best performing feature subset. It iteratively creates models and determines the best or the worst performing feature at each iteration. It constructs the subsequent models with the left features until all the features are explored. It then ranks the features based on the order of their elimination. In the worst case, if a dataset contains N number of features RFE will do a greedy search for 2N combinations of features.

```python
from sklearn.feature_selection import RFE
lm = LinearRegression()
scores = []
for i in range(1, x_train.shape[1]):
    rfe = RFE(lm, i)
    rkf = RepeatedKFold(n_splits = 10, n_repeats = 3, random_state = 1)
    score = cross_val_score(rfe, x_train, y_train, cv=rkf)
    scores.append(np.mean(score))
plt.plot(range(1, x_train.shape[1]), scores)   
```

```python
rfe = RFE(lm, 4)
rfe.fit(x_train, y_train)
rfe.support_
```

```python
rfe.score(x_test, y_test)
```

##### Embeded method
Embedded methods are iterative in a sense that takes care of each iteration of the model training process and carefully extract those features which contribute the most to the training for a particular iteration. Regularization methods are the most commonly used embedded methods which penalize a feature given a coefficient threshold.

This is why Regularization methods are also called penalization methods that introduce additional constraints into the optimization of a predictive algorithm (such as a regression algorithm) that bias the model toward lower complexity (fewer coefficients).

Examples of regularization algorithms are the LASSO, Elastic Net, Ridge Regression, etc.

```python
from sklearn.linear_model import Ridge
rg = Ridge(alpha=1.0)
rg.fit(x_train, y_train)
```

```python
print(rg.coef_, "\n", rg.intercept_, "\n", rg.score(x_test, y_test))
```

###### Ridge regression with build in cross validation

```python
from sklearn.linear_model import RidgeCV
rkf = RepeatedKFold(n_splits = 10, n_repeats = 3, random_state = 0)
rcv = RidgeCV(alphas = np.arange(0, 1, 0.1), cv=rkf)
rcv.fit(x_train, y_train)
```

```python
print(rcv.alpha_, "\n", rcv.coef_, "\n", rcv.intercept_, "\n", rcv.score(x_test, y_test))
```

###### Lasso

###### Elastic Net

```python
from sklearn.linear_model import ElasticNet
elasticnet = ElasticNet(alpha = 1.0, l1_ratio = 0.5)
elasticnet.fit(x_train, y_train)
```

```python
print(elasticnet.intercept_, "\n", elasticnet.coef_, "\n", elasticnet.score(x_test, y_test))
```

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.model_selection import RepeatedKFold
rkf = RepeatedKFold(n_splits = 10, n_repeats = 3, random_state = 0)
elnetCV = ElasticNetCV(alphas = np.arange(0, 1, 0.05), l1_ratio = np.arange(0, 1, 0.05), cv=rkf)
elnetCV.fit(x_train, y_train)
```

```python
print(elnetCV.alpha_, "\n", elnetCV.l1_ratio_, "\n", elnetCV.intercept_, "\n", elnetCV.coef_, "\n", elnetCV.score(x_test, y_test))
```

```python
y_predict = elnetCV.predict(x_test)
plt.plot(y_test, y_predict, ".")
x = np.linspace(0, 330, 100)
y = x
plt.plot(x, y)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print(mean_absolute_error(y_test, y_predict), "\n\n", mean_squared_error(y_test, y_predict), "\n\n", r2_score(y_test, y_predict))
```
